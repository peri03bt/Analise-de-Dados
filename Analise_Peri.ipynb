Note que **['minuto']**:
*  dtype=object
*  temos que organizar (**seperar 1 e 2 tempo**) e dá um `.value_counts()`

### Solução

```python
def convertendo_minutos(minutos):
    # Esta função recebe um valor de minutos no formato de string, como '45+1' ou '30'
    if '+' in minutos:
        # Se o sinal '+' estiver presente na string, isso significa que há acréscimo
        base_minutos, add_minutos = map(int, minutos.split('+'))
        # A função split('+') divide a string em duas partes: '45' e '1'
        # A função map(int, ...) converte essas partes em números inteiros
        return base_minutos + add_minutos / 10
        # Calcula o valor total de minutos considerando o acréscimo em décimos
    else:
        # Se não houver acréscimo, podemos converter diretamente a string em um inteiro
        return int(minutos)
```

```python
# Cria uma nova coluna no DataFrame aplicando a função convert_minutes à coluna 'minuto'
df_gols['minuto_convertido'] = df_gols['minuto'].apply(convertendo_minutos)
```

```python
df_gols['minuto_convertido'].unique()
```

```python
df_gols.head(3)
```

```python
gols_primeiro_tempo = df_gols[df_gols['minuto_convertido'] <= 45]['minuto_convertido'].value_counts().sum()
gols_segundo_tempo = df_gols[df_gols['minuto_convertido'] > 45]['minuto_convertido'].value_counts().sum()

print(f"Total de gols no primeiro tempo: {gols_primeiro_tempo}")
print(f"Total de gols no segundo tempo: {gols_segundo_tempo}")
```

### Gráfico

```python
labels = '1º Tempo', '2º Tempo'
sizes = gols_primeiro_tempo, gols_segundo_tempo
explode = (0, 0.1)
colors = sns.color_palette('Set3')

fig, ax = plt.subplots()
patches, texts, autotexts = ax.pie(sizes, labels=labels,
       explode=explode,
       autopct='%1.1f%%',
       pctdistance=.75,
       labeldistance=.1,
       shadow=True,
       startangle=90,
       colors=['skyblue', 'blue'])

# Personalizar a fonte e cores do texto
for text in texts:
    text.set_fontsize(12)
    if text.get_text() == '2º Tempo':
        text.set_fontweight('bold')
        text.set_color('white')
        
# Personalizar a fonte e cor do texto com os valores percentuais
for autotext in autotexts:
    autotext.set_fontsize(12)
    if autotext.get_text() == '59.2%':
        autotext.set_color('white')

    
plt.title('Distribuição de Gols no 1º e 2º Tempo', color='white')
plt.legend(labels=['3255', '4731'], 
           title="N° de Gols",
           loc='upper right',
           bbox_to_anchor=(1.2, 1)) # Adiconando um deslocamento extra para não ficar em cima do gráfico


# Define a cor do plano de fundo como transparente
fig.patch.set_facecolor('none')

plt.show()
```

## *2* - Gols por região;

### Análise

```python
conhecendo_df(df_regiao)
```

```python
df_regiao_limpo = df_regiao[['mandante_Placar', 'visitante_Placar', 'mandante_Estado', 'visitante_Estado']]
```

```python
conhecendo_df(df_regiao_limpo)
```

### Solução

```python
# Filtrar as linhas para o time mandante e visitante de cada estado
gols_mandante = df_regiao_limpo.groupby('mandante_Estado')['mandante_Placar'].sum()
gols_visitante = df_regiao_limpo.groupby('visitante_Estado')['visitante_Placar'].sum()

# Somar os gols feitos por cada estado (mandante e visitante)
total_gols_por_estado = gols_mandante + gols_visitante

print(gols_mandante)
print(gols_visitante)
print(total_gols_por_estado)

```

```python
# Ordenar a série em ordem decrescente
df_regiao_gols_sorted = total_gols_por_estado.sort_values(ascending=False)

df_regiao_gols_sorted
```

### Gráfico

```python
# Criar o gráfico de barras horizontais
plt.figure(figsize=(10, 6), facecolor='none')

# Configuração do fundo e bordas do gráfico
plt.gca().set_facecolor('none')  # Remover fundo branco
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_color('white')
plt.gca().spines['bottom'].set_color('white')

ax = df_regiao_gols_sorted.plot(kind='barh', facecolor='skyblue', edgecolor='none')

# Configuração da cor e estilo do texto
plt.yticks(fontweight='bold', color='white', fontsize=14)
plt.xticks(fontweight='bold', color='white')

plt.xlabel('')
plt.ylabel('')
plt.title('Total de Gols por Estado', fontsize=20, fontweight='bold', color='white')

# Adicionar legenda com número de gols
for i, gols in enumerate(df_regiao_gols_sorted):
    state = df_regiao_gols_sorted.index[i]
    color = 'yellow' if state == 'PE' else ('red' if state == 'SP' else 'white')
    ax.text(gols + 10, i, f'{gols}', va='center', fontsize=12, fontweight='bold', color=color)


# Configuração da cor e estilo do texto nos eixos
for tick_label in plt.gca().get_yticklabels():
    if tick_label.get_text() == 'PE':
        tick_label.set_color('yellow')
    elif tick_label.get_text() == 'SP':
        tick_label.set_color('red')
    else:
        tick_label.set_color('white')
    
# Remover quadrado que fecha o gráfico
plt.box(False)

plt.show()
```
